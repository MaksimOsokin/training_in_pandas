У нас есть список из 10 000 с разной суммой которые мы получили в прошлом. 

НЕобходимо узнать сколько денег было бы потрачено если на покупки свыше 8000 , мы бы выплачивали кэшбэг 5%.

```python
# Узнаем количество оплат покупок стоимостью выше 8 0000
# Завели счетчик что бы посмотреть количество оплат свыше 8000
counter = 0
for i in payments:
    if i >=8000:
       counter +=1
counter
```

```python
# Задача
# Нам прислали в одном списке данные о цене и валюте, в которой указана цена(рубли, доллоры)
# Неоходимо приобразовать все цены в рубли. 

usd_rate = 76.65
prices = [[1000,'RUR'], [4007, 'USD'], [5000,'RUR'], [2000, 'USD'], [4407, 'USD'], [26700, 'USD'], [4407, 'USD']]
# Посмотрим что содержит наш список с вложенным списком  
# print(prices[0][1])
# Создадим новый список 
new_pries = []
for i in prices:
    if i[1] != 'RUR':
        ner_pries = round((i[0]*usd_rate), 2)
        new_pries.append(ner_pries)
    else:
        new_pries.append(i[0])

prices = new_pries

print("Цены в рублях нового прайса", prices, "Общая сумма в рублях", sum(prices), sep="\n" )
```

```python
# Вспомню о методе преврающим строку в список SPLIT()

comments = 'клиент maks@mail.ru готов к оплате'

ls_comments = comments.split(' ')

mail = ''
for word in ls_comments:
    if word.find('@') > 0:
        mail = word
        
print(mail)
```

```python
# существуют и другие типы данных тюплы, в отличии от списков тюплы не получится перезаписать. 

# Еще один тип данных - МНОЖЕСТВА - это неупорядоченные колекции данных в которых каждое значение должно быть уникальным.
# Множества удобны для анализа спсисков. МЫ МОЖЕМ ПРЕОБРАЗОВАТЬ В МНОЖЕСТВО С ПОМОЩЮ set()

# Словари - хранят данные в формате (ключ : значение)

#  Ключь - Любой тип неизменяемых данных: числа , строки, кортежи , булин
#  Значение - вообще любой тип дан, в том числе другой словарь.


#  Создам словарь, где ключами являются строки а значениями - функция приведения к типу данных.  

type_functions_disct = {'строка'    : str,
                        'целое'     : int,
                        'дробное'   : float,
                       'логическое' : bool,
                       'список'     : list,
                       'кортеж'     : tuple,
                       'множество'  : set,
                       }


# Можем получить значение по ключу 

# type_functions_disct['список']

# Можем использовать значения словаря 
some_var = '4'
type(type_functions_disct['целое'](some_var))

# Словари можно использовать для создания баз данных скрин будет после текста

# В ловари легко добавлять данные 
type_functions_disct['словарь'] = dict 
# теперь можем обновить значение в словаре 
type_functions_disct['словарь'] = 'нечто'
```
